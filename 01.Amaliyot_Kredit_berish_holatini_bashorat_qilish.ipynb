```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
loan_dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data Scines Mohirdev/Datasets/loan_datasets.csv')
loan_dataset
```

```python
loan_dataset = loan_dataset.dropna()
```

```python
loan_dataset.isna().sum()
```

```python
loan_dataset
```

```python
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
loan_dataset
```

```python
loan_dataset['Dependents'].value_counts()
```

```python
loan_dataset['Dependents'].replace('3+', 4 , inplace=True)
```

```python
sns.countplot(x='Education',data=loan_dataset, hue = 'Loan_Status')
```

```python
sns.countplot(x='Married',data=loan_dataset, hue = 'Loan_Status')
```

```python
loan_dataset.replace({'Married':{'No':0, 'Yes':1},
                      'Gender': {'Male':1, 'Female':0},
                      'Education' : {'Graduate':1, 'Not Graduate':0},
                      'Self_Employed' : {'No':0, 'Yes':1},
                      'Property_Area' : {'Urban':0, 'Semiurban':1, 'Rural':2}}, inplace=True)
```

```python
loan_dataset
```

```python
x = loan_dataset.drop(columns=['Loan_ID','Loan_Status'], axis=1)
y = loan_dataset['Loan_Status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train, y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuray  = accuracy_score(x_train_prediction, y_train)
```

```python
print("Accuracy data score  :", training_data_accuray)
```

```python
x_test_prediction = classifier.predict(x_test)
test_data_accuray  = accuracy_score(x_test_prediction, y_test)
```

```python
print("Accuracy data score  :", test_data_accuray)
```
